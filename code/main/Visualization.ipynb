```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
t = open('TFIDFMatrix.txt')
lines = t.readlines()

items = []
for line in lines:
    print(line)
    print("SKLDJHF")
    items.append(line.split(" "))

#print(lines)
#print("SDKLUFHJKDLSJG")

print(items)
```

```python
print(np.shape(items))
```

```python
i = 0
for doc in items:
    print(i)
    doc.remove('\n')
    if (0<=i<=7):
        doc.append("airline")
    if (8<=i<=15):
        doc.append("disease")
    if (16<=i<=23):
        doc.append("bank")
    i+=1
print(items)
```

```python
print(np.shape(items))
```

```python
doc_df = pd.DataFrame(items)
```

```python
print(doc_df)
feature_list = doc_df.columns.values.tolist()
feature_list.remove(2291)
print(feature_list)
```

```python
vals = doc_df.loc[:, feature_list].values
print(vals)
#vals = StandardScaler().fit_transform(vals)
```

```python
print(vals)
print(doc_df[2291])
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(vals)

reduced_df = pd.DataFrame(data=components, columns=["X", "Y"])

reduced_df_final = pd.concat([reduced_df, doc_df[2291]], axis=1)
print(reduced_df_final)
print(np.shape(reduced_df_final))
```

```python
plt.scatter(reduced_df['X'], reduced_df['Y'])
plt.xlim(-0.2,0.1)
plt.ylim(-0.15,0.15)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('X', fontsize = 15)
ax.set_ylabel('Y', fontsize = 15)
ax.set_title('2-component PCA', fontsize = 20)

targets = ['airline', 'disease', 'bank']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = reduced_df_final[2291] == target
    ax.scatter(reduced_df_final.loc[indicesToKeep, 'X']
               , reduced_df_final.loc[indicesToKeep, 'Y']
               , c = color
               , s = 50)
#plt.xlim(-0.2,0.1)
#plt.ylim(-0.15,0.15)
ax.legend(targets)
ax.grid()
```
